```python
import pandas as pd
import os
import csv
```

```python
file = "hmeq.csv"
install_file = pd.read_csv(file)
install_file.head()
```

```python
install_file_sort = install_file.sort_values("VALUE",ascending = True)
install_file_sort = install_file_sort.rename(columns = {"LOAN": "Loan Total","MORTDUE":"Mortgage Due","VALUE": "Home Value",
                                                        "REASON":"Reason of Default","JOB":"Occupation",
                                                       "YOJ":"Years at Job","DEROG":"Major derogatory",
                                                       "DELINQ":"Delinquent credit lines",
                                                       "CLAGE":"Age Of Oldest Credit Line in Months",
                                                       "NINQ":"Number Of Recent Credit Inquiries", "CLNO":"Number of Credit Lines"
                                                       ,"DEBTINC":"Debt-to-Income Ratio"})
install_file_clean = install_file_sort.dropna(subset = ["Home Value"])
install_file_clean.head()
```

```python
#Cleans the dataframe to only those that default 
install_file_Default = install_file_clean[install_file_clean["BAD"] == 1]
Table1 = install_file_Default.loc[:,["BAD","Home Value"]]

bins = [0,20000,40000,60000,80000,100000,200000,400000,600000,800000,1000000]
group_names = ["<$20000","$20,000-39,999","$40,000-59,999","$60,000-79,999","$80,000-99,999","$100,000-199,999","$200,000-399,999","$400,000-599,999","$600,000-799,999","$800,000+"]

Table1["Home Value"] = pd.cut(Table1["Home Value"],bins,labels=group_names,right=False)
NewTable = Table1.groupby("Home Value").count()
NewTable
```

```python
#Cleans the dataframe to only those that did not default 
install_file_not_Default = install_file_clean[install_file_clean["BAD"] == 0]
Table1 = install_file_not_Default.loc[:,["BAD","Home Value"]]

bins = [0,20000,40000,60000,80000,100000,200000,400000,600000,800000,1000000]
group_names = ["<$20000","$20,000-39,999","$40,000-59,999","$60,000-79,999","$80,000-99,999","$100,000-199,999","$200,000-399,999","$400,000-599,999","$600,000-799,999","$800,000+"]

Table1["Home Value"] = pd.cut(Table1["Home Value"],bins,labels=group_names,right=False)
NewTable = Table1.groupby("Home Value").count()
NewTable
```
