1)  The goal is to obtain income tax data by county & zip code, and provide the average income by zip code or city, to be merged with team members data on default and housing costs.  Most county zip codes are issued numerically, however, due to suburb expansion, several county zip codes are out of sequence.  For this reason, we oculd not simply list or sort our IRS tax data by a zip code range.

As an option, I decided to look up CA income tax data, and it turned out to be available on a csv file from 1992 to 2010, by zip code and was much cleaner to obtain the necessasry data.
4)  The ca state income tax data was imported as a pandas df, cleaned and munged to display income data by zip code. The only difficulty was figuring out that we had to convert the gross income to 'type float' in order to display the data we needed.

6)  The ca state income tax data with added mean income column, was now sorted based on the year value.  Each year from 2010 - 2017 was entered as its own df.
7)  The df's are now to be merged with team memners datasets.

```python
# Import dependencies, pandas & numpy to calc income mean by zip code
import pandas as pd
import numpy as np
```

```python
# reference csv file to import tax data
file = 'CA_Income_Tax_Stats_By_Zip_Code.csv'

# read csv
ca_zipcode_income_tax = pd.read_csv('SoCal Zip Code Income 2010.csv') 
ca_zipcode_income_tax.head()
```

```python
# Get a list of all of our columns for easy reference
ca_zipcode_income_tax.columns
```

```python
# Extract data to show only "ID", "Taxable Year", "Zip Code", "Retruns", "CA AGI",

reduced_zipcode_df = ca_zipcode_income_tax.loc[:, ["ID", "Taxable Year", "Zip Code", "City", "Returns", "CA AGI"]]
reduced_zipcode_df.head()
```

```python
# check number of rows & columns
reduced_zipcode_df.shape
```

```python
caagi = reduced_zipcode_df['CA AGI']
caagi
```

```python
returnsss = reduced_zipcode_df['Returns']
returnsss
```

```python
reduced_zipcode_df["CA AGI"].str.replace(",","").astype(float) / reduced_zipcode_df["Returns"].str.replace(",","").astype(float)
```

```python
# add a column for mean income (CA AGI/Returns)
reduced_zipcode_df['Avg_AGI'] = reduced_zipcode_df["CA AGI"].str.replace(",","").astype(float) / reduced_zipcode_df["Returns"].str.replace(",","").astype(float)
reduced_zipcode_df
```

```python
reduced_zipcode_df_a = reduced_zipcode_df.sort_values("Taxable Year",ascending = False)
reduced_zipcode_df_a = reduced_zipcode_df_a[reduced_zipcode_df_a["Taxable Year"]>= 2010]
reduced_zipcode_df_a.head()
```

```python
#Renaming Columns
reduced_zipcode_df_a = reduced_zipcode_df_a.rename(columns = {"Taxable Year":"Year", "Avg_AGI":"Avg Income"})
#Rounding Avg Income
reduced_zipcode_df_a["Avg Income"] = reduced_zipcode_df_a["Avg Income"].apply(lambda x: round(x,2))
reduced_zipcode_df_a
```

```python
# have to sort data by year, 2010 to 2017, use:  df.loc[df['column_name'] == some_value]
zip_inc_2017 = reduced_zipcode_df_a.loc[reduced_zipcode_df_a['Year'] == 2017]
zip_inc_2017.head()
```
